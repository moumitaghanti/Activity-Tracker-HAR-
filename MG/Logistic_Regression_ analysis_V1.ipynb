```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
activity_data = pd.read_csv("Outputs/output_file_acc.csv")
activity_data.head()
activity_data.Profile = pd.Categorical(activity_data.Profile)
activity_data['Profile_code'] = activity_data.Profile.cat.codes
activity_data.Activity = pd.Categorical(activity_data.Activity)
activity_data['Activity'] = activity_data.Activity.cat.codes

clean_data = activity_data.drop("Profile", axis=1)
clean_data = clean_data.drop("Date_time", axis=1)

clean_data["Profile_code"].unique()
#activity_data["Profile"].unique()
clean_data.describe()
clean_data.dtypes
clean_data.dropna()
```

```python
# Assign X (data) and y (target)
X = clean_data.drop("Activity", axis=1)
y = clean_data["Activity"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True,
          intercept_scaling=1, max_iter=100, multi_class='ovr', n_jobs=1,
          penalty='l2', random_state=None, solver='liblinear', tol=0.0001,
          verbose=0, warm_start=False)
```

```python
X_train.head()
```

```python
 classifier.fit(X_train, y_train)
```

```python
 LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True,
          intercept_scaling=1, max_iter=100, multi_class='ovr', n_jobs=1,
          penalty='l2', random_state=None, solver='liblinear', tol=0.0001,
          verbose=0, warm_start=False)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
predictions = classifier.predict(X_test)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```
