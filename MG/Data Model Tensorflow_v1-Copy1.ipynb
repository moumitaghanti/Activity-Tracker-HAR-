```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from scipy import stats
```

```python
WISDM_data = pd.read_csv("Outputs/WISDM.csv",header=None)

WISDM_data.columns = ["Profile_ID","Activity","ts","X","Y","Z","A","B","C","D","E"]
WISDM_data.head()
WISDM_df = WISDM_data[["X","Y","Z","ts","Activity","Profile_ID"]]
WISDM_df.head()
```

```python
activity_data = pd.read_csv("Outputs/output_file_acc.csv")
activity_data.head()
```

```python
new_data = pd.read_csv("Outputs/output_file_acc_ROC_v2.csv")
new_data.head()
```

```python
new_df = new_data[["ts","X","Y","Z","Profile","Activity"]]

new_df.Profile = pd.Categorical(new_df.Profile)
new_df['Profile'] = new_df.Profile.cat.codes
new_df.Activity = pd.Categorical(new_df.Activity)
new_df['Activity'] = new_df.Activity.cat.codes
new_df.head()
```

```python

activity_data.Profile = pd.Categorical(activity_data.Profile)
activity_data['Profile_code'] = activity_data.Profile.cat.codes

clean_data = activity_data.drop("Profile", axis=1)
clean_data = clean_data.drop("Date_time", axis=1)
clean_data["Profile_ID"] = clean_data["Profile_code"]+36
clean_data = clean_data.drop("Profile_code", axis=1)

clean_data.head()
```

```python
actdict = {"stand":"Standing",
            "sit":"Sitting",
            "stairsu":"Upstairs",
            "stairsd":"Downstairs",
            "walk":"Walking",
            "laying":"Laying"
           }

Act_text = ["stand","sit", "laying", "walk", "stairsu", "stairsd"]
for act in Act_text:
    clean_data.loc[clean_data['Activity'] == act, 'Activity'] = actdict[act]
clean_data.head()
```

```python

new_df = new_df[1:]
new_df["Xr"]= new_df["Xr"].astype('float32')
```

```python
def feature_normalize(dataset):
    mu = np.mean(dataset,axis = 0)
    sigma = np.std(dataset,axis = 0)
    return (dataset - mu)/sigma
new_df.dropna(axis=0, how='any', inplace= True)
new_df['X'] = feature_normalize(new_df['X'])
new_df['Y'] = feature_normalize(new_df['Y'])
new_df['Z'] = feature_normalize(new_df['Z'])
```

```python
new_df["Profile"].unique()
```

```python
df_test = new_df[new_df['Profile'] > 5]
df_train = new_df[new_df['Profile'] <= 5]
#df_traindf_train = clean_data
```

```python
x_train_df = df_train[["X","Y","Z","ts","Profile"]]
x_train_df 
X_train = x_train_df.values
X_train
```

```python
y_train_df = df_train["Activity"]
y_train_df
y_train = to_categorical(y_train_df)

X_train = X_train.astype('float32')
y_train = y_train.astype('float32')
y_train[:10]
```

```python
 # Read the testing data
X_test_df = df_test[["X","Y","Z","ts","Profile"]]
X_test = X_test_df.values
X_test.shape
```

```python
# Read the testing labels
y_test_df = df_test["Activity"]
# One-hot encode the integer labels
y_test = to_categorical(y_test_df)
y_test.shape

```

```python
# Create an empty sequential model
model = Sequential()
```

```python
# Add the first layer where the input dimensions are the 561 columns of the training data
model.add(Dense(100, activation='relu', input_dim=X_train.shape[1]))
```

```python
model.add(Dense(100, activation='relu'))
```

```python
# The output layer has 13 columns that are one-hot encoded
y_train.shape
```

```python
model.add(Dense(100, activation='relu'))
```

```python
model.add(Dense(100, activation='relu'))
```

```python
model.add(Dense(100, activation='relu'))
```

```python
# Add output layer
model.add(Dense(y_train.shape[1], activation="softmax"))
```

```python
 # Compile the model using categorical_crossentropy for the loss function, the adam optimizer,
# and add accuracy to the training metrics
model.compile(loss="categorical_crossentropy",
              optimizer="adam", metrics=['accuracy'])
```

```python
# Use the training data to fit (train) the model
model.fit(
    X_train,
    y_train,
    epochs=100,
    shuffle=True,
    verbose=2
)
```

```python
# Save the model
model.save("smartphone_trained1.h5")
```

```python
# Load the model
from tensorflow.keras.models import load_model
model = load_model("smartphone_trained1.h5")
```

```python
# Evaluate the model using the training data
model_loss, model_accuracy = model.evaluate(X_test, y_test, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
