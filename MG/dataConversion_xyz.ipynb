```python
 # Import Dependencies
import pandas as pd
import numpy as np
```

```python
csv_path = "z.csv"
z_df = pd.read_csv("z.csv",header=None)
z_df.columns = ["Date_time", "Z"]
z_df.head()
y_df = pd.read_csv("y.csv", header=None)
y_df.columns = ["Date_time", "Y"]
y_df.head()
x_df = pd.read_csv("x.csv",
                  header=None)
x_df.columns = ["Date_time", "X"]
x_df.head()
```

```python
xy_df = x_df.merge(y_df, on="Date_time")
xyz_df = xy_df.merge(z_df, on="Date_time")
xyz_df["Profile"] = "MG"
xyz_df["Activity"] = "standing"
xyz_df["Date_time"]= pd.to_datetime(xyz_df["Date_time"])
xyz_df['ts'] = xyz_df[['Date_time']].apply(lambda x: x[0].timestamp(), axis=1).astype(int)
xyz_df
```

```python
xyz_df.to_csv("sample_mg.csv")
```
