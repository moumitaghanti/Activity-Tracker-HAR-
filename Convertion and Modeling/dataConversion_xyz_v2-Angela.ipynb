```python
 # Import Dependencies
import pandas as pd
import numpy as np
```

```python
output_df = []

Prof_text = "ar"
Act_text = ["stand","sit", "laying", "walk", "stairsu"]

for act in Act_text:
    csv_path = "Outputs/"+act+"_"+Prof_text+"_acc.csv"
    print(csv_path)
    try:
        df  = pd.read_csv(csv_path)
        
        df["Profile"] ="AR"
        df["Activity"]= act
        #df["Date_time"] = df["Time_stamp"]
        clean_df = df[["Date_time","X","Y","Z","Profile","Activity"]]
        clean_df["Date_time"]= pd.to_datetime(clean_df["Date_time"])
        clean_df['ts'] = clean_df[['Date_time']].apply(lambda x: x[0].timestamp(), axis=1).astype(int)
        out_path = "Outputs/"+act+"_"+"AR"+"_acc.csv"
        clean_df.to_csv(out_path)
          
    except FileNotFoundError:
        print(csv_path+" file format is missing")
        
clean_df
```
